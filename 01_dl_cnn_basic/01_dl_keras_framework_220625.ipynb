## **Graduate admission prediction**

### **1. Import libraries**

```python
import warnings
warnings.filterwarnings('ignore')
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'

import numpy as np
import pandas as pd
import tensorflow as tf
```

### **2. Load the dataset**

```python
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/update/dl_cnn/1_dl_cnn_basic/data/gpascore.csv')
data
```

### **3. Explore the data**

```python
data.isnull().sum()
```

```python
data.dropna(how='any', inplace=True)
```

```python
data.isnull().sum()
```

### **4. Preprocess the data**

```python
y = data['admit'].values
y[:10]
```

```python
x = []
x_data = data.drop('admit', axis=1)
for i, rows in x_data.iterrows():
    x.append(rows.values)
x[:10]
```

### **5. Build/Compile/Train the model**

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(64, activation='tanh'),
    tf.keras.layers.Dense(128, activation='tanh'),
    tf.keras.layers.Dense(1, activation='sigmoid'),
])
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(np.array(x), y, epochs=1000)
```

### **6. Make predictions**

```python
pred = model.predict([[800, 4.00, 3], [600, 3.00, 1]])
print(np.round(pred, 4))
```
